```python
from sklearn.metrics import precision_recall_curve, auc
import matplotlib.pyplot as plt
```

```python
pr_aucs = []
roc_aucs = []

for i in range(1, 6):

    # Load fold data
    train = pd.read_csv(f"train_smote_{i}.csv")
    test  = pd.read_csv(f"test_{i}.csv")

    # Train model
    model = ydf.RandomForestLearner(
        label="Target",
        num_trees=500
    ).train(train)

    # Evaluate on test set
    eval_result = model.evaluate(test)

    # Extract PR-AUC
    pr_auc = eval_result.characteristics[0].pr_auc
    roc_auc = eval_result.characteristics[0].roc_auc

    print(f"Fold {i} PR-AUC: {pr_auc:.4f}")
    print(f"Fold {i} ROC-AUC: {roc_auc:.4f}")
    
    if pr_auc is not None and not np.isnan(pr_auc):
        pr_aucs.append(pr_auc)
    else:
        print(f"Skipping fold {i} — PR-AUC undefined")

    
    if roc_auc is not None and not np.isnan(roc_auc):
        roc_aucs.append(roc_auc)
    else:
        print(f"Skipping fold {i} — ROC-AUC undefined")    

# Summary
    print("\n==============================")
    print(f"Mean PR-AUC: {np.mean(pr_aucs):.4f}")
    print(f"Mean ROC-AUC: {np.mean(roc_aucs):.4f}")
    print(f"Std Dev PR-AUC: {np.std(pr_aucs):.4f}")
    print("==============================")

```
